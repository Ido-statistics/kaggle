```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

## データの確認

```python
#データ読み込み
df = pd.read_csv("ks-projects-201801.csv")
#データの中身確認
display(df.head(10))
display(df.describe())
print(df.dtypes)
```

## 変数の整理

ID ⇨要らない 

pledged / usd pledged / usd_pledged_real / backers ⇨クラウドファンディング後のものなので予測に使うのは違う
goal ⇨usd_goal_realで良さそうなので多分要らない

```python
df = df.drop(["ID", "backers", "goal", "pledged", "usd pledged", "usd_pledged_real"], axis=1)
display(df.head())
```

```python
 #目的変数について
df['state'].unique()
```

```python
#成功or失敗なので、failedとsucessfulだけ取る
df = df[df.state.str.contains("successful")|df.state.str.contains("failed")]
df['state'].unique()
```

```python
# succesfulを1, failedを0に変化
df.replace({"successful":1, "failed":0}, inplace=True)
display(df)
```

 ## 日付変数の変換
 deatlineとlaunchedの差分より経過時間(日単位)を算出し、説明変数とした.

```python
# deadline列及びlaunched列のTimestanp型への変換
df["deadline"] = pd.to_datetime(df["deadline"])
df["launched"] = pd.to_datetime(df["launched"])

# 開始日から締切日までの期間を日単位で差分を計算
df["days"] = pd.Series((df["deadline"] - df["launched"]).dt.days)

# deadline列及びlaunched列を削除
df = df.drop(["deadline", "launched"], axis=1)
display(df.head())
```

## 残りのカテゴリ変数をダミー変数にする
nameでget_dummiesを使用したらなぜかプログラムが動かなくなってしまったため、nameは文字数に変換して扱う。

```python
#name以外
dummy = ['category','main_category','currency','country']
df = pd.get_dummies(df, columns=dummy)

display(df.head())
```

```python
#nameについて
# name列の欠損値が含まれる行の確認
df[df["name"].isnull()]
```

```python
#name列に欠損値が含まれている場合、それを削除
df = df.dropna(subset=['name'],axis=0)
```

```python
#name列を文字数に変換した行を追加する
df['name_length'] = df['name'].apply(lambda x:len(x))
display(df.head())
```

```python
#name列の削除
df = df.drop('name',axis=1)
display(df.head())
```

## 訓練データとテストデータに分ける

```python
#目的変数と説明変数にわける
label = df["state"]
data = df.drop("state",axis=1)
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=0.3)
```

```python
#訓練用データを標準化する
data_train["usd_goal_real"] = (data_train["usd_goal_real"] - data_train["usd_goal_real"].mean()) / data_train["usd_goal_real"].std(ddof=0)
data_train["days"] = (data_train["days"] - data_train["days"].mean()) / data_train["days"].std(ddof=0)
data_train["name_length"] = (data_train["name_length"] - data_train["name_length"].mean()) / data_train["name_length"].std(ddof=0)
```

```python
data_train
```

```python
y_train = label_train.values
X_train = data_train.values
```

## 特徴量選択
ステップワイズ法を用いた

```python
from sklearn.linear_model import SGDClassifier
from sklearn.feature_selection import RFECV

estimator = SGDClassifier(loss='log',random_state=1234)

# RFECVは交差検証によってステップワイズ法による特徴選択を行う
# cvにはFold（=グループ）の数，scoringには評価指標を指定する
rfecv = RFECV(estimator,cv=2,scoring='accuracy')

#fitで選択を実行
rfecv.fit(X_train,y_train)
```

```python
# 削除すべき特徴の名前を取得 
removed_idx  = ~rfecv.get_support()
# 削除してもよい特徴の名前を取得する
remove_feature = data_train.columns[removed_idx]
remove_feature
```

```python
# drop関数で特徴を削除
selected_X = data_train.drop(remove_feature, axis=1)
selected_X
```

```python
selected_X_train = selected_X.values
```

## テスト用データの作成

```python
#テスト用データを標準化する
data_test["usd_goal_real"] = (data_test["usd_goal_real"] - data_test["usd_goal_real"].mean()) / data_test["usd_goal_real"].std(ddof=0)
data_test["days"] = (data_test["days"] - data_test["days"].mean()) / data_test["days"].std(ddof=0)
data_test["name_length"] = (data_test["name_length"] - data_test["name_length"].mean()) / data_test["name_length"].std(ddof=0)
```

```python
#特徴量選択をしたテスト用データ
remove_feature_test = data_test.columns[removed_idx]
selected_X2 = data_test.drop(remove_feature_test, axis=1)
selected_X2
```

```python
selected_X_test = selected_X2.values
y_test = label_test.values
```

## モデル作成

```python
X_train = selected_X_train
X_test = selected_X_test
```

## AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix,classification_report
```

```python
#訓練誤差
clf = AdaBoostClassifier(RandomForestClassifier(max_depth=8,
                                                min_samples_leaf=3,
                                                min_samples_split=2,
                                                random_state=1234,
                                                criterion="gini"),
                         n_estimators=50, random_state=1234)
clf.fit(X_train, y_train)
print("score=", clf.score(X_train, y_train))
```

```python
#汎化誤差
# テストデータに対する予測を実行
y_pred_test = clf.predict(X_test)

# 正答率を計算
accuracy =  accuracy_score(y_test, y_pred_test)
print('正答率（Accuracy） = {:.3f}%'.format(100 * accuracy))

# Precision, Recall, F1-scoreを計算
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred_test)

# Precision, Recall, F1-scoreを表示
print('適合率（Precision） = {:.3f}%'.format(100 * precision[0]))
print('再現率（Recall） = {:.3f}%'.format(100 * recall[0]))
print('F1値（F1-score） = {:.3f}%'.format(100 * f1_score[0]))
```
